```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu_results = np.load('results.npy')

plt.figure(figsize=(20, 14))
k_session = 12
k_point = 10
for i in range(k_session):
    plt.subplot(3, 4, i+1)
    plt.scatter(range(k_point), mu_results[i], c='Red', marker='*')
    plt.plot(range(k_point), np.ones(k_point)*(0.75 + 0.05*i))
    plt.ylim(0.7, 1.4)
    plt.ylabel('$\mu$')
    plt.xlabel('experiments')
    plt.title(f'$\mu$ = {0.75 + 0.05*i}', fontsize=8)
    plt.grid()

# plt.show()
plt.savefig('mu_experiment.png')
```
